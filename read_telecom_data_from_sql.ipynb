# Read data from sql

```python
%matplotlib inline
from sqlalchemy import create_engine
import json
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python

db_credentials = json.loads(open('database_credentials.json').read())
```

```python
cred_string = ('mysql+pymysql://' +
               db_credentials['username'] +
               ':' +
               db_credentials['user_pass'] +
               '@' +
               db_credentials['db_host'] +
               '/' +
               db_credentials['db_name']
              )

conn = create_engine(cred_string)
conn.table_names()
```

```python
query = 'select * from telecom_churn_data'
data = pd.read_sql(query,conn)
```

```python
print(data.shape)
data.head()
```

```python
columns = [
'State', 'Account Length', 'Area Code', 'Phone',
'International Plan',
'VMail Plan', 'VMail Message',
'Day Mins', 'Day Calls', 'Day Charge',
'Eve Mins', 'Eve Calls', 'Eve Charge',
'Night Mins', 'Night Calls', 'Nights Charge',
'International Mins', 'International Calls', 'International Charge',
'CustServ Calls',
'Churn'
]
data.rename(columns={f'columns{i}': j for i, j in zip(range(1, 22), columns)}, inplace=True) 
```
